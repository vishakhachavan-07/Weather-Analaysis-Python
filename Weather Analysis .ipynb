```python
import pandas as pd 
```

| Line | Code                           | Meaning                                                        |
| ---- | ------------------------------ | -------------------------------------------------------------- |
| 1    | `data = {}`                    | Creates a dictionary named `data` to store information.        |
| 2    | `'Day': ['Mon', 'Tue', 'Wed']` | Creates a column called **Day** with values Mon, Tue, Wed.     |
| 3    | `'Temp': [39, 38, 37]`         | Creates a column called **Temp** with temperatures 39, 38, 37. |
| 4    | `pd.DataFrame(data)`           | Converts the dictionary into a table (DataFrame).              |
| 5    | `df = pd.DataFrame(data)`      | Stores the table in a variable named **df**.                   |
| 6    | `print(df)`                    | Displays the table on the screen.                              |


```python
data ={ 'Day': ['Mon', 'Tue', 'Wed'],
        'Temp': [39,38,37]
       }
df = pd.DataFrame(data)
print(df)
```

| Part | Code                    | Meaning                                                  |
| ---- | ----------------------- | -------------------------------------------------------- |
| 1    | `df['Temp']`            | Selects the **Temp** column from the DataFrame.          |
| 2    | `df['Temp'].idxmax()`   | Finds the index (row number) of the highest temperature. |
| 3    | `df.loc[...]`           | Retrieves the complete row at that index.                |
| 4    | `hottest =`             | Stores that row in a variable called `hottest`.          |
| 5    | `print("Hottest Day:")` | Prints the heading.                                      |
| 6    | `print(hottest)`        | Prints the hottest day's information.                    |


```python
hottest = df.loc[df['Temp'].idxmax()]

print("hottest Day:")
print(hottest)
```

| Part | Code                    | Meaning                                                  |
| ---- | ----------------------- | -------------------------------------------------------- |
| 1    | `df['Temp']`            | Selects the **Temp** column from the DataFrame.          |
| 2    | `df['Temp'].idxmin()`   | Finds the index (row number) of the lowest temperature. |
| 3    | `df.loc[...]`           | Retrieves the complete row at that index.                |
| 4    | `hottest =`             | Stores that row in a variable called `hottest`.          |
| 5    | `print("Coldest Day:")` | Prints the heading.                                      |
| 6    | `print(hottest)`        | Prints the Coldest day's information.                    |


```python
Coldest = df.loc[df['Temp'].idxmin()]

print("Coldest day")
print(Coldest)
```

```python
| Step | Code             | Meaning                                                      |
| ---- | ---------------- | ------------------------------------------------------------ |
| 1    | `df['Temp']`     | Selects the **Temp** column from the DataFrame.              |
| 2    | `.mean()`        | Calculates the average (mean) of all values in that column.  |
| 3    | `average_temp =` | Stores the average value in a variable named `average_temp`. |
| 4    | `print()`        | Displays the result on the screen.                           |
```

```python
average_temp = df['Temp'].mean()
print("Average Temperature =", average_temp)
```

```python
import pandas as pd

data = {
    'Day' : ['MONDAY', 'TUESDAY', 'WEDNESDAY', 'THURSDAY', 'FRIDAY', 'SATURDAY', 'SUNDAY'],
    'Temp': [39, 37, 38, 35, 30, 28, 36]
}

df = pd.DataFrame(data)

print("weather data")
print(df)

Hottest = df.loc[df['Temp'].idxmax()]
Coldest = df.loc[df['Temp'].idxmin()]
Avg_Temp = df['Temp'].mean()

print("\nHottest day:")
print(Hottest)


print("\nColdest  day:")
print(Coldest)


print("\nAverage Temperature:", Avg_Temp)

```
